```python
import pandas as pd
import numpy as np
```

```python
def _read_meta(meta_path):
    meta = pd.read_csv(meta_path).fillna(-1)
    meta.loc[:, "SubBrandId"] = meta.SubBrandId.fillna(-1).astype(np.int64)
    return meta
```

```python
meta = _read_meta("/data2/datasets/clobotics/ccth/indices/ccth_sku_metadata.csv")
meta120 = _read_meta("/data2/datasets/clobotics/ccth/indices/ccth_sku_metadata_0120.csv")
print(len(meta), len(meta120))
```

### Differences and relationship

- in terms of ProductId coverage, meta is pure subset of meta120
- for product_id 1052167, the subbrand_id in meta is 1052231, while it's 1063802 in meta120. Seems like it's an error in meta120.
